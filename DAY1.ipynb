 # agar dots ke color ko control karna ho ,size ko control karna ho to kya karenge 

```python
colours=np.random.rand(50) # yaha random numbers ko generate kia hai colours varible mei 
sizes=100*np.random.rand(50) # yaha 100 se multiply karne kaa matlab hai ki size badh gaya hai data points kaa 
plt.scatter(x,y,c=colours,s=sizes,alpha=0.4) # alpha intensity ko represent karta hai 
plt.xlabel("x axis")
plt.ylabel("y label")
plt.title("random data")
```

# How to use 3 d cordinate

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
plt.show()
```
